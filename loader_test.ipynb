```python
import matplotlib.pyplot as plt

from torch.utils.data import  DataLoader
from pathlib import Path

from loader.dataloader import SphericalProjectionKitti

def test_loader(loader, iter = 1):
    i = 0
    for data in loader:
        _, proj_points, labels, proj_labels, _  = data
        print(proj_points.shape, labels.shape, proj_labels.shape)
        plt.figure(figsize=(20,80))
        plt.imshow(proj_labels[0], cmap="gray")

        plt.figure(figsize=(20,80))
        plt.imshow(proj_points[0][2], cmap="gray")
        i+=1
        if i == iter:
            break


```

```python
test_data = SphericalProjectionKitti(Path("examples/data/kitti/00/"), length=1)

test_loader_basic = DataLoader(test_data, batch_size=1, shuffle=False)

test_loader(test_loader_basic)
```
